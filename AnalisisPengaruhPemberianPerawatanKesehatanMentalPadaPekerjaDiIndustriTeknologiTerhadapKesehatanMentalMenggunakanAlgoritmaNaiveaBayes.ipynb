### Analisis Pengaruh Pemberian Perawatan Kesehatan Mental pada Pekerja di Industri Teknologi terhadap Kesehatan Mental menggunakan Algoritma Naive Bayes (Studi kasus: Pekerja Industri Teknologi di Amerika Serikat)


```python
import pandas as pd
import numpy as np
```

```python
# input data
mentalhealth=pd.read_excel("DataModelling-survey-mental-health-on-tech-workplaces.xlsx")
# Menampilkan data
mentalhealth.head()
```

```python
# menampilkan informasi data
mentalhealth.info()
```

```python
# Mengecek apakah ada deret yang kosong
mentalhealth.empty
```

```python
# Melihat ukuran dari data
mentalhealth.size
```

```python
# Variabel independen
x = mentalhealth.drop(["treatment"], axis = 1)
x.head()
```

```python
# Variabel dependen
y = mentalhealth["treatment"]
y.head()
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

```python
# Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
# Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
modelnb = GaussianNB()
# Memasukkan data training pada fungsi klasifikasi naive bayes
nbtrain = modelnb.fit(x_train, y_train)
nbtrain.class_count_
```

```python
# Menentukan hasil prediksi dari x_test
y_pred = nbtrain.predict(x_test)
y_pred
```

```python
# Menentukan probabilitas hasil prediksi
nbtrain.predict_proba(x_test)
```

```python
# import confusion_matrix model
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Menghitung nilai akurasi dari klasifikasi naive bayes 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
